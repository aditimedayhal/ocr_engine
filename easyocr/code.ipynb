```python
import cv2
import easyocr
import numpy as np
from tkinter import Tk, filedialog

# Initialize EasyOCR Reader
reader = easyocr.Reader(['en'])

# Global variables for mouse callback
ref_points = []
cropping = False

def select_roi(event, x, y, flags, param):
    global ref_points, cropping

    # Mouse button pressed down
    if event == cv2.EVENT_LBUTTONDOWN:
        ref_points = [(x, y)]
        cropping = True

    # Mouse button released
    elif event == cv2.EVENT_LBUTTONUP:
        ref_points.append((x, y))
        cropping = False

        # Draw rectangle around selected ROI
        cv2.rectangle(image_copy, ref_points[0], ref_points[1], (0, 255, 0), 2)
        cv2.imshow("Select Region", image_copy)

# Select an image file using file dialog
def choose_image():
    root = Tk()
    root.withdraw()  # Hide the root window
    image_path = filedialog.askopenfilename(
        title="Select an Image",
        filetypes=[("Image Files", "*.jpg *.jpeg *.png *.bmp *.tiff")]
    )
    return image_path

# Choose image dynamically
image_path = choose_image()
if not image_path:
    print("No image selected. Exiting...")
    exit()

# Load the selected image
image = cv2.imread(image_path)
image_copy = image.copy()

cv2.namedWindow("Select Region")
cv2.setMouseCallback("Select Region", select_roi)

print("Select the region of interest (ROI) by clicking and dragging.")
print("Press 'c' to perform OCR on the selected area or 'q' to quit without OCR.")

while True:
    cv2.imshow("Select Region", image_copy)
    key = cv2.waitKey(1) & 0xFF

    # Press 'c' to perform OCR on the selected ROI
    if key == ord("c"):
        if len(ref_points) == 2:
            # Crop the selected region
            roi = image_copy[ref_points[0][1]:ref_points[1][1], ref_points[0][0]:ref_points[1][0]]

            # Convert to RGB format for EasyOCR
            roi_rgb = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)

            # Perform OCR
            results = reader.readtext(roi_rgb)
            print("OCR Results:")
            for (bbox, text, prob) in results:
                print(f"Text: {text}, Probability: {prob:.2f}")

            # Display the cropped region
            cv2.imshow("Selected ROI", roi)
            cv2.waitKey(0)
        else:
            print("Error: Select a valid region first.")

    # Press 'q' to quit
    elif key == ord("q"):
        break

cv2.destroyAllWindows()
```
